```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
```

```python
data_path = "/Users/samwwong/Downloads/mb_t1_t2_4groups.xlsx"
df = pd.read_excel(data_path)
```

```python
drop_columns = ['seg_id', 'molec_id', 'dx_date', 'age', 'alive', 'os', 'pfs']
df = df.drop(drop_columns, axis = 1)
df['sex'] = df['sex'].replace({'M': 0, 'F': 1})
df = df[~df['molecular'].isin(['3 or 4', 'shh or 3 or 4', '?'])].reset_index(drop = True)

df.head()
```

```python
X = df.drop('molecular', axis = 1)
y = df['molecular']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```
